## ข้อ 1

```python
import pandas as pd
df=pd.read_csv('Auto-MPG.csv')
```

## ข้อ 2

```python
temp=df['horsepower'].mode()[0]
df['horsepower'] = df['horsepower'].replace('?', temp)
```

## ข้อ 3

```python
temp=df['name']
for i in range(398):
    temp2=temp[i].split()
    df.at[i, 'name']=temp2[0]
df['name']=df['name'].replace({'chevy': 'chevrolet', 'vw': 'volkswagen'})
df
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['name']=label_encoder.fit_transform(df['name'])
df
```

## ข้อ 4

```python
y=df['mpg']
x=df.drop(columns=['mpg'])
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(-1,1))
x=pd.DataFrame(scaler.fit_transform(x), columns=x.columns)
x
```

```python
from sklearn.model_selection import train_test_split
xTrain,xTest,yTrain,yTest=train_test_split(x,y,test_size=0.2,random_state=42)
```

## ข้อ 5

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(xTrain,yTrain)
```

```python
print('w1: ',model.coef_)
print('w0: ',model.intercept_)
```

```python
y_pred =model.predict(xTest)
y_pred
```

## ข้อ 6

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
print("MAE: ",mean_absolute_error(yTest,y_pred))
print("MSE: ",mean_squared_error(yTest,y_pred))
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y') 
df.set_index('Date', inplace=True)

newDf = pd.DataFrame()
newDf['w(t)'] = df['P2'].resample('W').sum()
newDf['w(t+1)'] = newDf['w(t)']
newDf['w(t-1)'] = newDf['w(t)']

newDf = newDf.reset_index()
newDf['Week'] = range(1, len(newDf) + 1)
newDf = newDf[['Week','w(t-1)', 'w(t)','w(t+1)']]
df=newDf
```

```python
df['w(t+1)']=df['w(t+1)'].shift(-1)
df['w(t-1)']=df['w(t-1)'].shift(+1)
df=df.dropna()
df
```

```python
y=df['w(t+1)']
x=df.drop(columns=['w(t+1)'])

scaler = StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x), columns=x.columns)
```

```python
xTrain=x[:99]
yTrain=y[:99]

xTest=x[99:]
yTest=y[99:]
```

```python
model=LinearRegression()
model.fit(xTrain,yTrain)
```

## ข้อ 12

```python
result=model.score(xTest,yTest)
print("accuracy: {:.2f}%".format(result*100))
```
